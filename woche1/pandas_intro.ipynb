## Eine Tabelle erstellen aus Listen
Bei einem Wettlauf gibt es drei Strecken: 15 km, 42 km und 60 km. F체r den Lauf verwenden wir 5 Listen, eine mit Vornamen der Teilnehmer, eine mit gew채hlter Streckenl채nge, eine mit Alter der Teilnehmer, eine mit Ortsangabe der Teilnehmer und eine Liste, die das Geschlecht der Teilnehmer festh채lt. 

```python
vorname=["Sari", "Raphael","Silvia","Charly"]
strecke=[15,42,60,42]
alter=[20,23,15,70]
postleitzahl=[6520,6045,8055,3994]
geschlecht=["w","m","w","m"]
```

Die f체nf Listen beinhalten f체nf Informationen 체ber jede an einem Lauf teilnehmende Person.  Man kann die gespeicherten Informationen auch als eine einzige Tabelle verstehen!


Um aber mit Tabellen zu arbeiten, muss man zuvor noch ein Paket installieren:
Lassen Sie die n채chsten beiden Zellen laufen.

```python
import pandas as  pd #Paket von Befehlen f체r die Datenverarbeitung
```

```python
datenLauf={"vorname" : ["Sari", "Raphael","Silvia","Charly"],
"strecke" : [15,42,60,42],
"alter" : [20,23,15,70],
"postleitzahl" : [6520,6045,8055,3994],
"geschlecht" : ["w","m","w","m"]}

dfLauf=pd.DataFrame(datenLauf)
dfLauf
```

Der Befehl `pd.DataFrame(datenLauf)` macht aus einer Sammlung von Daten eine Tabelle, einen so genannten DataFrame. Dazu benutzt es einen Befehl aus der pd Liberary. Meist wird kurz df f체r dataFrame benutzt.

**Wichtig zu wissen:** Die Art und Weise wie wir die Daten eingegeben haben, `datenLauf={ "Spaltenname": [Liste], ...}` nennt man ein Dictionary. Dictionaries werden in Python auch sonst h채ufig verwendet, wir schauen uns diese Datenstruktur aber nicht genauer an. Benutzen Sie jetzt einfach die Syntax, ohne genauer deren Zweck zu analysieren.

**Exkurs** Man kann auch eine weitere Zeile hinzuf체gen, aber um die Bezeichnungen zu verstehen, m체ssen wir unsere Daten zuerst analysieren und einige Befehle lernen.


```python
dfLauf.loc[len(dfLauf)] = ['Jan', 42, 27, 8032, 'm']  #Hinzuf체gen
print(dfLauf)
dfLauf.drop(dfLauf.index[len(dfLauf)-1], inplace= True) #Wieder l철schen
print(dfLauf)
```
